```r
#Ex 0

plot(0:15,dbinom(0:15,10,0.3),col="red",main = "Binomial Distribution")
points(0:15,dbinom(0:15,5,0.45),col="blue")
points(0:15,dbinom(0:15,7,0.7),col="green")
legend(x="topright", y="topright",legend=c("(10,03)","(5,0.45)","(7,0.7)"),cex=1,col=c("red","green","blue"),pch=1)


plot(0:15,pbinom(0:15,10,0.3),col="red",,main = "Cumulative Binomial Distribution")
points(0:15,pbinom(0:15,5,0.45),col="blue")
points(0:15,pbinom(0:15,7,0.7),col="green")
legend(x="bottomright", y="bottomright",legend=c("(10,03)","(5,0.45)","(7,0.7)"),cex=1,col=c("red","green","blue"),pch=1)


plot(0:20,dpois(0:20,3),col="red",main = "Poisson Distribution")
points(0:20,dpois(0:20,5),col="blue")
points(0:20,dpois(0:20,7),col="green")
legend(x="topright", y="topright",legend=c("lambda = 3","lambda = 5","lambda = 7"),cex=1,col=c("red","green","blue"),pch=1)


plot(0:20,ppois(0:20,3),col="red",,main = "Cumulative Poisson Distribution")
points(0:20,ppois(0:20,5),col="blue")
points(0:20,ppois(0:20,7),col="green")
legend(x="bottomright", y="bottomright",legend=c("lambda = 3","lambda = 5","lambda = 7"),cex=1,col=c("red","green","blue"),pch=1)

```

```r
#Ex 1

data <- data.frame(x=c("x","p1","p2"),"x1" = c(15.58,0.15,0.14), "x2" = c(15.9,0.21,0.05), "x3" = c(16,0.35,0.64), "x4" = c(16.1,0.15,0.08), "x5" = c(16.2,0.14,0.09))
E1 <- sum(data[1,-1]*data[2,-1])
E2 <- sum(data[1,-1]*data[3,-1])
V1 <- sum(data[1,-1]*data[2,-1]*data[1,-1]) - sum(data[1,-1]*data[2,-1])^2
V2 <- sum(data[1,-1]*data[3,-1]*data[1,-1]) - sum(data[1,-1]*data[3,-1])^2 

message("The mean value for probability set 1 is ", E1)
message("The mean value for probability set 2 is ", E2)
message("The variance for probability set 1 is ", V1)
message("The variance for probability set 2 is ", V2)
```

```r
sample <- rexp(50,1/30)

hist(sample,main="Waiting time of 50 patients")
prob_10_min <- pexp(10,1/30)
m <- mean(sample)
diff <- m - 30
prob_1_h <- 1 - pexp(60,1/30) 

message("The probability of a waiting time of less than 10 min is ", prob_10_min)
message("The probability of a waiting time of more than 1 hour is ", prob_1_h)
message("The difference between the mean of the sample and of the population is ", diff)
```
